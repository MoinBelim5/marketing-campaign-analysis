```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("marketing_campaign_data.csv")
```

```python
# Check Data Types
print(df.dtypes)
```

```python
# Convert Start_Date, End_Date into datetime type
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
```

```python
df['End_Date'] = pd.to_datetime(df['End_Date'])
```

```python
print(df.dtypes)
```

```python
# Campaign duration in days
df['Campaign_Duration'] = (df['End_Date'] - df['Start_Date']).dt.days

# Click-Through Rate (CTR)
df['CTR'] = np.where(df['Impressions'] > 0,
(df['Clicks'] / df['Impressions']).round(2), 0)

# Conversion Rate
df['Conversion_Rate'] = np.where(df['Clicks'] > 0,
(df['Conversions'] / df['Clicks']).round(2), 0)

# ROI
df['ROI'] = np.where(df['Budget (USD)'] > 0,
((df['Revenue (USD)'] - df['Budget (USD)']) / df['Budget (USD)']).round(2),0)
```

```python
df
```

```python
# Top Performing Campaigns by ROI
top_roi = df.sort_values(by='ROI', ascending=False).head(10)
print(top_roi[['Campaign_Name', 'ROI', 'Revenue (USD)', 'Budget (USD)']])
```

```python
# CTR by Target Age Group
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Target_Age_Group', y='CTR', estimator='mean', errorbar=None)
plt.title('Click-Through Rate by Age Group')
plt.tight_layout()
plt.show()
```

```python
# Revenue over Time
df_by_date = df.groupby('Start_Date').agg({'Revenue (USD)': 'sum'}).reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(data=df_by_date, x='Start_Date', y='Revenue (USD)')
plt.title('Revenue Trend Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Revenue & Conversions by Region
region_summary = df.groupby('Region').agg({
    'Revenue (USD)': 'sum',
    'Conversions': 'sum'
}).reset_index()

print(region_summary)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=region_summary, x='Region', y='Revenue (USD)')
plt.title('Revenue by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CTR & Conversion Rate by Channel
channel_summary = df.groupby('Channel').agg({
    'CTR': 'mean',
    'Conversion_Rate': 'mean',
    'ROI': 'mean'
}).reset_index().round(2)

print(channel_summary)
```

```python
## AVG CTR by Channel
plt.figure(figsize=(10,5))
sns.barplot(data=channel_summary, x='Channel', y='CTR')
plt.title('Average CTR by Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Campaign Duration vs Conversions
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Campaign_Duration', y='Conversions', hue='Channel', alpha=0.6)
plt.title('Campaign Duration vs Conversions')
plt.xlabel('Campaign Duration (Days)')
plt.ylabel('Conversions')
plt.tight_layout()
plt.show()
```

```python
# KPI Summary Table
print("Average CTR:", df['CTR'].mean().round(2))
print("Average Conversion Rate:", df['Conversion_Rate'].mean().round(2))
print("Average ROI:", df['ROI'].mean().round(2))
print("Average Campaign Duration:", df['Campaign_Duration'].mean().round(2))
```

```python
## Campaign Classification
df['Campaign_Status'] = np.where(df['ROI'] >= 0, 'Profitable', 'Loss Making')
## df['Campaign_Status']
```

```python
df['Revenue_per_Dollar'] = (df['Revenue (USD)'] / df['Budget (USD)']).round(2)
## df['Revenue_per_Dollar']
```

```python
## Budget efficiency bucket

df['Efficiency_Bucket'] = pd.cut(
    df['ROI'],
    bins=[-float('inf'), 0, 1, float('inf')],
    labels=['Loss', 'Break-even', 'High Return']
)
## df['Efficiency_Bucket']
```

```python
## Top & Bottom Campaign comparison

top_10 = df.sort_values('ROI', ascending=False).head(10)
bottom_10 = df.sort_values('ROI').head(10)

compare_cols = [
    'ROI', 'CTR', 'Conversion_Rate', 
    'Budget (USD)', 'Campaign_Duration'
]

top_10[compare_cols].mean(), bottom_10[compare_cols].mean()
```

```python
## Channel x ROI matrix

channel_roi = (
    df.groupby('Channel')['ROI']
    .mean().round(2)
    .sort_values(ascending=False)
)
channel_roi
```
